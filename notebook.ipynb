### Data Preparation

```python
# Keeping only the features having more than a 1% correlation score with PSS_score
columns_to_keep = ['PSS_score', 'Conscientiousness', 'Extraversion', 'Neuroticism', 'sleep_time', 'sleep_duration', 'PSQI_score', 'num_calls', 'num_sms', 'skin_conductance', 'mobility_distance']
df = df[columns_to_keep]

# Renaming columns
df.columns = df.columns.str.lower()
df.columns
```

```python
from sklearn.model_selection import train_test_split
df_full_train, df_test = train_test_split (df, test_size = 0.20, random_state = 10)
df_train, df_val = train_test_split (df_full_train, test_size = 0.25, random_state = 10)
len(df_train), len(df_val), len(df_test)
```

```python
y_train = df_train["pss_score"]
y_val = df_val["pss_score"]
y_test = df_test["pss_score"]
del df_train["pss_score"]
del df_val["pss_score"]
del df_test["pss_score"]
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse = False)
```

```python
dict_train = df_train.to_dict(orient = "records")
dict_val = df_val.to_dict(orient = "records")
dict_test = df_test.to_dict(orient = "records")
```

### Linear regression model

```python
# Let's train a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

lr = LinearRegression()
X_train = dv.fit_transform(dict_train)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_train)
rmse_score = root_mean_squared_error(y_train, y_pred)
print(rmse_score)

```

```python
lr.coef_
```

```python
X_val = dv.transform(dict_val)
y_pred = lr.predict(X_val)
rmse_score = root_mean_squared_error(y_val, y_pred)
print(rmse_score)
```

### Regularization of the linear regression model

#### Using Ridge Regression

```python
from sklearn.linear_model import Ridge
for r in [0.001, 0.01, 0.1, 1, 10, 100]:
    ridge_model = Ridge(alpha = r)
    ridge_model.fit(X_train, y_train)
    y_pred = ridge_model.predict(X_val)
    rmse_score = root_mean_squared_error(y_val, y_pred)
    print(f'rmse score for alpha value {r} is {round(rmse_score,4)}')
```

Ridge option does not provide better RMSE.

#### Using Lasso Regression

```python
from sklearn.linear_model import Lasso
for r in [0.001, 0.01, 0.1, 1, 10, 100]:
    lasso_model = Lasso(alpha = r)
    lasso_model.fit(X_train, y_train)
    y_pred = lasso_model.predict(X_val)
    rmse_score = root_mean_squared_error(y_val, y_pred)
    print(f'rmse score for alpha value {r} is {round(rmse_score,3)}')
```

Compared to Ridge regression, Lasso regression gives slightly better RMSE score but it's not a really interesting enhancement of the model.

#### Using Elastic Net

```python
from sklearn.linear_model import ElasticNet
alphas = [0.001, 0.01, 0.1, 1, 10, 100]
l1_ratio = [0.001, 0.01, 0.1, 0.5, 1]
rmse_min = 9999
alpha_best = None
l1_best = None
for alpha in alphas:
    for l1 in l1_ratio:
        elasticnet_model = ElasticNet(alpha=alpha, l1_ratio=l1)
        elasticnet_model.fit(X_train, y_train)
        y_pred = elasticnet_model.predict(X_val)
        rmse_score = root_mean_squared_error(y_val, y_pred)
        if rmse_score < rmse_min:
            rmse_min = rmse_score
            alpha_best = alpha
            l1_best = l1
        print(f'rmse score for alpha value {alpha} and l1_ratio value {l1}  is {round(rmse_score,3)}')
print(f'\n The best rmse score for alpha value {alpha_best} and l1_ratio value {l1_best}  is {round(rmse_min,3)}')
```

The best rmse score (8.574) with Elastic Net regression is not better than the best model using the Lasso (alpha = 1 , rmse = 8.576). For this, we will use only the Lasso regression model.

### KFold Cross-Validation on Lasso regression model with tuning

```python
def train(df_train, y_train, C=1.0):
    dicts = df_train.to_dict(orient='records')

    dv = DictVectorizer(sparse=False)
    X_train = dv.fit_transform(dicts)

    model = Lasso(alpha = C)
    model.fit(X_train, y_train)
    
    return dv, model
```

```python
def predict(df, dv, model):
    dicts = df.to_dict(orient='records')

    X = dv.transform(dicts)
    y_pred = model.predict(X)

    return y_pred
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=1)
alphas = [0.1, 0.2, 0.5, 0.75, 1, 5, 10, 20]

columns = list(df_full_train.columns)
columns.remove("pss_score")

for alpha in alphas :
    scores = []
    for train_idx, val_idx in kfold.split(df_full_train):
        df_train_kfold = df_full_train.iloc[train_idx]
        df_val_kfold = df_full_train.iloc[val_idx]

        y_train_kfold = df_train_kfold["pss_score"]
        y_val_kfold = df_val_kfold["pss_score"]

        dv, model = train(df_train_kfold[columns], y_train_kfold, C=alpha)
        y_pred = predict(df_val_kfold[columns], dv, model)

        rmse_score = root_mean_squared_error(y_val_kfold, y_pred)
        scores.append(rmse_score)

    print('alpha=%s %.3f +- %.3f' % (alpha, np.mean(scores), np.std(scores)))
```

For a linear regression model, the best Lasso regression model we got with an mean rmse score of 8.598 is with alpha = 0.5 (case kfold of 5 splits).

The RMSE value of this decision tree model (8.679) is not better than the best score of the linear regression model (8.598).

### Random Forest model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=10, random_state= 1, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
rmse_score = root_mean_squared_error(y_val, y_pred)
print(rmse_score)
```

with this RF model, the RMSE score (9.131) is even worse than before. <br>
What could be the causes ? <br> 
- n_estimators
- Max_depth parameter
<br> Next, we will try to tune the two parameters.

### Tuning the random forest model

```python
# Tunning the n_estimators
scores = []
for n in range(10,201,10):
    rf = RandomForestRegressor(n_estimators=n, random_state= 1, n_jobs=-1)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    score = root_mean_squared_error(y_val, y_pred)
    scores.append([n, score])
columns = ["n_estimators","rmse"]
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
import matplotlib.pyplot as plt
plt.plot(df_scores.n_estimators, df_scores.rmse)
```

With n_estimators of 110 and 200, we reach local minimas for RMSE value. 

```python
# Tunning the n_estimators and max_depth parameters
scores = []
d_values = [1, 2, 3 , 5, 10, 15, 20]
for d in d_values:
    for n in range(10,201,10):
        rf = RandomForestRegressor(n_estimators=n, max_depth=d, random_state= 1, n_jobs=-1)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_val)
        score = root_mean_squared_error(y_val, y_pred)
        scores.append([d, n, score])
columns = ["max_depth", "n_estimators","rmse"]
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
df_scores.groupby('max_depth').rmse.agg('mean')
```

With the Random forest model, we did get a slightly better score (8.594) than with the linear model (8.598). 

### Using XGBoost 

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label = y_train, feature_names = features)
dval = xgb.DMatrix(X_val, label = y_val, feature_names = features)
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
def parse_xgb_output(output, parameter_value):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((parameter_value, it, train, val))
    
    columns = ['parameter', 'num_iter', 'train_rmse', 'val_rmse']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```

```python
%%capture output

eta_value = 0.05

xgb_params = {
    'eta': eta_value, 
    'max_depth': 3,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, num_boost_round=100,
                  evals=watchlist)
```

```python
parse_xgb_output(output, eta_value).sort_values("val_rmse",ascending=True)
```

```python
y_pred = model.predict(dval)
root_mean_squared_error(y_val, y_pred)
```
